```python

def caculatesundays(startyear,startmonth,startday,endyear,endmonth,endday):
    datelist=[]
    before=datetime.datetime(startyear,startmonth,startday)
    after=datetime.datetime(endyear,endmonth,endday)
    rr = rrule.rrule(rrule.WEEKLY,byweekday=relativedelta.SU,dtstart=before)
    s=rr.between(before,after,inc=True)
    for d in s:
        if int(d.month/10)==0:
            mon=str(0)+str(d.month)
        else:
            mon=str(d.month)
        if int(d.day/10)==0:
            da=str(0)+str(d.day)
        else:
            da=str(d.day)
        datelist.append((str(d.year)+'-'+mon+'-'+da))
    return datelist    
    
          #get all sundays between two dates
    
```

```python
def getbooklist(datelist,booktype):
    for pubdate in datelist:
        r = requests.get("https://api.nytimes.com/svc/books/v3/lists/"+pubdate+"/"+booktype+".json?&api-key="+os.getenv('auth_key'))
        writejson(pubdate,booktype,r.json()) 
    








```
