```python

def caculatesundays(startyear,startmonth,startday,endyear,endmonth,endday):
    datelist=[]
    before=datetime.datetime(startyear,startmonth,startday)
    after=datetime.datetime(endyear,endmonth,endday)
    rr = rrule.rrule(rrule.WEEKLY,byweekday=relativedelta.SU,dtstart=before)
    s=rr.between(before,after,inc=True)
    for d in s:
        if int(d.month/10)==0:
            mon=str(0)+str(d.month)
        else:
            mon=str(d.month)
        if int(d.day/10)==0:
            da=str(0)+str(d.day)
        else:
            da=str(d.day)
        datelist.append((str(d.year)+'-'+mon+'-'+da))
    return datelist    
    
          #get all sundays between two dates
    
```

```python
def readjson(path):                                               #read json
    with open(path,'r',encoding='utf-8',errors='ignore') as df:
        data=json.load(df)
        return data
```

```python
allfiction=[]
for files in glob.glob(os.path.dirname(os.getcwd())+"/raw data/fiction-booklist/*.json"):
    booklist=readjson(files)
    for book in booklist["results"]["books"]:
        a=book
        info=book["title"]+"--"+book["author"]
        allfiction.append(info)
    
    
    
```

```python

dic=FreqDist(allfiction)

timeonlist=sorted(dic.items(),key=lambda x:x[1],reverse=True)
timeonlist
```

```python
bestsellerlist=[]
for auth in timeonlist:
    a1=auth[0].split("--")[0]
    a2=auth[0].split("--")[1]
    a1=a1.replace(",","_")
    a2=a2.replace(",","_") 
    bestsellerlist.append((a1,a2,auth[1]))
    
    
bestsellerlist
```
