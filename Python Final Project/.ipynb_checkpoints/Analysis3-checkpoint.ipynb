```python
def readjson(path):                                               #read json
    with open(path,'r',encoding='utf-8',errors='ignore') as df:
        data=json.load(df)
        return data
```

```python
def cleanword(rawcontent):
    wordlist=[i for i in word_tokenize(rawcontent.lower()) if i not in stopwords.words('english') and i not in string.punctuation and i.isalpha()]
    return wordlist     # clean review and get word list
```

```python
def getstarreview(path,star):
    wordlist=[]
    dic=readjson(path)["reviews"]
    for eachrest in dic:
        if eachrest["rating"]==star:
            wordlist.extend(cleanword(eachrest["text"]))
    return wordlist        
            
        
    # get 5,4,3,2,1 star review
    
```

```python
def getstarwords(pathlist,star):
    wordslist=[]
    for path in pathlist:
        wordslist.extend(getstarreview(path,star))
    return wordslist    
    
```

```python
list5star=getstarwords(reviewpath,5)
list4star=getstarwords(reviewpath,4)
list3star=getstarwords(reviewpath,3)
list2star=getstarwords(reviewpath,2)
list1star=getstarwords(reviewpath,1)


#  store word list into list by review star
```

```python
raw_data = {'Review_Rating': ['5Star', '4Star', '3Star', '2Star', '1Star'],
        'place': [502, 320, 125, 53, 37],
        'food': [398, 253, 119, 65, 44],
        'service': [158, 81, 69, 34, 21]}
df_review = pd.DataFrame(raw_data, columns = ['Review_Rating', 'place', 'food', 'service'])
df_review     #   output dataframe
```

```python

pos = list(range(len(df_review['place'])))
width = 0.25


fig, ax = plt.subplots(figsize=(10,5))

# Create a bar with pre_score data,

plt.bar(pos,
        
        df_review['place'],
        
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#EE3224',
        
        label=df_review['Review_Rating'][0])


plt.bar([p + width for p in pos],
       
        df_review['food'],
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#F78F1E',
        
        label=df_review['Review_Rating'][1])


plt.bar([p + width*2 for p in pos],
   
        df_review['service'],
        # of width
        width,
        # with alpha 0.5
        alpha=0.5,
        # with color
        color='#FFC222',
     
        label=df_review['Review_Rating'][2])

# Set the y axis label
ax.set_ylabel('Score')

# Set the chart's title
ax.set_title('Restaurant concerning views about place,food,service')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df_review['Review_Rating'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(df_review['place'] + df_review['food'] + df_review['service'])] )


plt.legend(['place', 'food', 'service'], loc='upper left')
plt.grid()
plt.show()




```
