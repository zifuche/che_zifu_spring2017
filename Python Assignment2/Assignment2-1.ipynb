```python

import nltk.corpus
from nltk.corpus import inaugural

```

```python
import glob
allpath=[]                     #save  all path
for name in glob.glob('/Users/zifuche/nltk_data/corpora/inaugural/*.txt'):
    allpath.append(name)
    
needpath=allpath[-25:-2]      #  12 txt
    
    
```

```python
print(needpath)
```

```python

from string import punctuation
def readtxt(path):
    wordlist=[]
    f=open(path,'r')
    for line in f.readlines():
        for word in line.split(" "):
            word=''.join(c for c in word if c not in punctuation)
            word=word.lower()
            if word.endswith("\n"):
                word=word[:-2]                
            if word!="\n" and word!=None and word!="":
                wordlist.append(word)                
    return wordlist   
        

```

```python
totalwordlist=[]
for path in needpath:
    l=readtxt(path)
    totalwordlist.extend(l)
    
    
    
    





```

```python
from nltk.probability import FreqDist
```

```python
freq=FreqDist(totalwordlist)
```

```python
w=sorted(freq.items(),key=lambda x:x[1],reverse=True)
print(w)
```

```python
def rank(w):
    ra=[]
    count=1
    for t in w:
        r=(count,t[0],t[1])     
        count=count+1
        ra.append(r)
    return ra
          
        
           
    
```

```python
csvlist=rank(w)

t=csvlist[0:500]
```
