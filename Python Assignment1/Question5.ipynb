# Question5
Use Gutenberg and Web_text data. Find out what are the top 5 words that Shakespeare used but we do not use in currently.
Take top 50 words from Shakespeare (all 3 books) and top 50 from Web_text (all the records).
Remove punctuation and stop words.
Remove the words we still use today, and get the unused list. Show the top 5 elements. 


```python
import nltk
```

```python
sha=['shakespeare-caesar.txt',
 'shakespeare-hamlet.txt',
 'shakespeare-macbeth.txt',]
```

```python
shawords=[]
for f in sha:
    for word in nltk.corpus.gutenberg.words(f):
        word=word.lower()
        if word.isalpha() and word not in nltk.corpus.stopwords.words('english'):
            shawords.append(word)
list1=shawords           
        
        
      

    
```

```python

sf={}

for word in shawords:
    if word not in sf:
        sf[word]=1
    else:
        sf[word]=sf[word]+1
        
     


w=sorted(sf.items(),key=lambda x:x[1],reverse=True) 
print(w[0:50])        
```

```python

sf2=[]

for word in nltk.corpus.webtext.words():
        word=word.lower()
        if word.isalpha() and word not in nltk.corpus.stopwords.words('english'):
            sf2.append(word)




 

list2=sf2
print(list2)
```

```python
cf={}
for word in list2:
    word=word.lower()
    if word not in cf and word.isalpha():
        cf[word]=1
    if word in cf and word.isalpha():
        cf[word]=cf[word]+1
        
print((sorted(cf.items(),key=lambda x:x[1],reverse=True)[0:50]) )       
       

```

```python

list3=[word for word in list1 if word not in list2]


print(list3)
```

```python
listtop={}
for word in list3:
    if word not in listtop:
        listtop[word]=1
    else:
        listtop[word]=listtop[word]+1
        
        
        
```

```python
top=sorted(listtop.items(),key=lambda x:x[1],reverse=True)
print(top)
```

```python
print(top[:5])         #answer
```
